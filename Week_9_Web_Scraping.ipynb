# Web Scraping

```python
import urllib.request
import os
from bs4 import BeautifulSoup
```

```python
print(os.getcwd())
# os.chdir('...')
```

```python
url = 'https://emojiisland.com/pages/free-download-emoji-icons-png'
req = urllib.request.urlopen(url)
```

```python
# print(dir(req))
req.__dict__
```

```python
html = ''
if req.status==200:   html = req.read()
print(len(html))
```

```python
soup = BeautifulSoup(html, 'html.parser')

# Retrieve all of the anchor tags

tags = soup('img')
os.makedirs('png', exist_ok=True)
```

```python
for tag in tags:
  
    imgurl = tag.get('src')
    if (not ('png' in imgurl)):
        continue
    try:
        outimgname = os.path.basename(imgurl).split('?')[0]
        if ('http' in imgurl):
            urllib.request.urlretrieve(imgurl, os.path.join('png', outimgname))
        else:
            urllib.request.urlretrieve('https:' + imgurl, os.path.join('png', outimgname))
    except:
        print(outimgname + ' Get Error.')    
        continue     
```

```python
# In addition
# Request Status
print(req.status)
```

```python
# Get the title of webpage: web_title
print(soup.title)
```
